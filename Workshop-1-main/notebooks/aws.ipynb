```python
import numpy as np
import matplotlib.pyplot as plt

# Stages of data center construction
stages = [
    "Site Selection & Design",
    "Permitting & Compliance",
    "Site Prep & Foundation",
    "Structural Construction",
    "Electrical & Cooling",
    "Server Installation",
    "Go Live"
]

# Durations in months (can use np.random or manually set)
durations = np.array([2, 3, 2, 5, 3, 2, 1])
start_months = np.cumsum(np.insert(durations[:-1], 0, 0))
colors = plt.cm.viridis(np.linspace(0, 1, len(stages)))
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

for i, (stage, start, duration) in enumerate(zip(stages, start_months, durations)):
    ax.barh(y=stage, width=duration, left=start, color=colors[i])

ax.set_xlabel("Months")
ax.set_title("Timeline of Data Center Construction")
plt.tight_layout()
plt.show()
```

```python
costs = [40, 25, 20, 15]
labels = ['Construction', 'Electrical & HVAC', 'Equipment', 'Security & Networking']

fig2, ax2 = plt.subplots()
ax2.pie(costs, labels=labels, autopct='%1.1f%%', startangle=90)
ax2.set_title("Cost Breakdown of Data Center Construction")
plt.show()
```

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'Stage': stages,
    'Start': start_months,
    'Duration': durations
})

fig = px.bar(df, x="Duration", y="Stage", orientation='h', color="Stage",
        title="Data Center Construction Timeline",
        labels={'Duration': 'Months'})
fig.show()
```

```python
fig.savefig("data_center_timeline.png", dpi=300)
fig2.savefig("cost_breakdown_pie.png", dpi=300)
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Define the construction phases and durations
stages = [
    "Site Selection & Design",
    "Permitting & Compliance",
    "Site Prep & Foundation",
    "Structural Construction",
    "Electrical & Cooling",
    "Server Installation",
    "Go Live"
]

# Durations in months
durations = np.array([2, 3, 2, 5, 3, 2, 1])
start_months = np.cumsum(np.insert(durations[:-1], 0, 0))

# Create a DataFrame
df = pd.DataFrame({
    'Stage': stages,
    'Start': start_months,
    'Duration': durations
})

# Plot with Plotly
fig = px.bar(
    df,
    x="Duration",
    y="Stage",
    orientation='h',
    color="Stage",
    title="Data Center Construction Timeline",
    labels={'Duration': 'Months'}
)

fig.update_layout(
    xaxis_title="Time (months)",
    yaxis_title="Construction Stage",
    showlegend=False,
    bargap=0.4
)

fig.show()
```
